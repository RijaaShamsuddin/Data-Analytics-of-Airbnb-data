## Task 2b: Data Cleaning (Python)

- Check for missing values in the dataframe and display the count in ascending order. **If the values are missing, impute the values as per the datatype of the columns.**
- Check whether there are any duplicate values in the dataframe and, if present, remove them.
- Display the total number of records in the dataframe before and after removing the duplicates.

```python
## Check for missing values in the dataframe and display the count in ascending order.
missing_values = df.isnull().sum()
sort_missing_values = missing_values.sort_values(ascending=True)
print(sort_missing_values)
```

```python
## Impute missing values based on the datatype of the columns
for column in df.columns:
    if df[column].dtype == object:
        df[column].fillna('None', inplace=True)  # For object columns, impute with None string
    elif df[column].dtype == 'float64':
        df[column].fillna(df[column].mean(), inplace=True)  # For float64 datatype columns, impute with the column mean
    elif df[column].dtype == 'int64':
        df[column].fillna(df[column].mean(), inplace=True)  # For float64 datatype columns, impute with the column mean

# Display the DataFrame after imputing missing values
print("\nDataFrame after imputing missing values:")
print(df)
```

```python
## Check for missing values have been imputed in the dataframe
missing_values_imputed = df.isnull().sum().sort_values(ascending=True)
print(missing_values_imputed)
```

```python
## Check whether there are any duplicate values in the dataframe and if present remove them.
duplicate_rows = df.duplicated()
print("Number of duplicate rows:", duplicate_rows.sum())

#remove duplicates
drop_duplicates = df.drop_duplicates()

#DataFrame after removing the duplicates
total_records_after_droping = drop_duplicates.shape[0]
print("Total number of records after removing duplicates:", total_records_after_droping)
```

```python
## Display the total number of records in the dataframe before removing the duplicates
total_records_before_droping = Airbnb_data.shape[0]
print("Total number of records before removing duplicates:",total_records_before_droping)
```

- Rename the column `availability 365` to `days_booked`

```python
## Rename the column.
rename_columns = df.rename(columns={'availability 365': 'days_booked'})
rename_columns.head(5)
```

```python
## Convert all column names to lowercase and replace the spaces with an underscore "_"
df.columns = df.columns.str.lower().str.replace(" ","_")
df.head(5)
```

```python
## Remove the dollar sign and comma from the columns. If necessary, convert these two columns to the appropriate data type.

#Check which columns contain the dollar sign
columns_with_dollar_sign = [
    col for col in df.columns 
    if df[col].apply(lambda x: isinstance(x, str) and '$' in x).any()
    ]
print("Columns with dollar sign:", columns_with_dollar_sign)
print(df.dtypes)

#Remove the dollar sign from float64 datatype columns only
df['price'] = df['price'].replace({'\$': '', ',': '','None':np.nan}, regex=True).astype(float)
df['service_fee'] = df['service_fee'].replace({'\$': '', ',': '','None': np.nan}, regex=True).astype(float)
df.columns
df.head(5)
```

### Task 4: Exploratory Data Analysis (Any Tool)

- List the count of various room types avaliable in the dataset.
- Which room type has the most strict cancellation policy?
- List the average price per neighborhood group, and highlight the most expensive neighborhood to rent from.

If using Python for this exercise, please include the code in the cells below. If using any other tool, please include screenshoots of your work.

```python
print(df.columns)
print(df.dtypes)
```

```python
## List the count of various room types avaliable with Airbnb
count_room_type = df['room_type'].value_counts()
print(count_room_type)
```

```python
## Which room type adheres to more strict cancellation policy
original_array = df.cancellation_policy.unique()
print(original_array)

value_to_remove = "None"

df_unique = original_array[original_array != value_to_remove]
print(df_unique)

#GroupBy room_type and cancellation_policy with df_unique values and alotting a column with the counts
grouped = df[df['cancellation_policy'].isin(df_unique)].groupby(['room_type', 'cancellation_policy']).size().reset_index(name='count')

print("\n") 
print(grouped)

#The room type with the most strict cancellation policy
most_strict_room_type = grouped[grouped['cancellation_policy'] == 'strict'].nlargest(1, 'count')
print("\n") 
print(most_strict_room_type)
```

```python
## List the average price per neighborhood group, and highlight the most expensive neighborhood to rent from.

# Filter out rows with "None(NaN)" in 'NeighborhoodGroup'
df_unique_neighbourhood_group = df[df['neighbourhood_group'] != 'None']

# Calculate the average price per neighborhood group
average_prices = df_unique_neighbourhood_group.groupby('neighbourhood_group')['price'].mean().reset_index()
print(average_prices)

# Identify the most expensive neighborhood
most_expensive_neighborhood = average_prices.loc[average_prices['price'].idxmax(), 'neighbourhood_group']
print("\nThe most expensive neighbourhood group is:", most_expensive_neighborhood)
```

```python
# Create a chart to highlight the most expensive neighbourhood

import altair as alt
# Calculate the average price per neighborhood group
average_prices = df_unique_neighbourhood_group.groupby('neighbourhood_group')['price'].mean().reset_index()

# Identify the most expensive neighborhood
most_expensive_neighborhood = average_prices.loc[average_prices['price'].idxmax(), 'neighbourhood_group']

# Create the bar chart using Altair
bar_chart = alt.Chart(average_prices).mark_bar().encode(
    x='neighbourhood_group',
    y='price',
    color=alt.condition(
        alt.datum.neighbourhood_group == most_expensive_neighborhood,
        alt.value('orange'),   # Highlighted color for the most expensive neighborhood
        alt.value('steelblue')  # Default color for other neighborhoods
    )
).properties(
    title='Average Price per Neighborhood Group',
    width=500,
    height=300
)

# Display the chart using the Altair viewer
display(bar_chart)
```

```python
##Create a horizontal bar chart to display the top 10 most expensive neighborhoods in the dataset.

avg_price_neighborhood = df.groupby('neighbourhood')['price'].mean().round(1).sort_values(ascending = False)
top10_exp_neighbourhoods = avg_price_neighborhood.head(10)
print(top10_exp_neighbourhoods)

top10_exp_neighbourhoods.plot(kind='barh', color = 'slateblue')
plt.xlabel('Price')
plt.ylabel('Neighborhood')
plt.title('Top 10 Most Expensive Neighborhoods')
plt.show()
```

```python
##Create another chart with the 10 cheapest neighborhoods in the dataset.

avg_price_neighborhood = df.groupby('neighbourhood')['price'].mean().round(1).sort_values()
top10_cheapest_neighbourhoods = avg_price_neighborhood.head(10)
print(top10_cheapest_neighbourhoods)

top10_cheapest_neighbourhoods.plot(kind='barh', color = 'teal')
plt.xlabel('Price')
plt.ylabel('Neighborhood')
plt.title('Top 10 Most Expensive Neighborhoods')
plt.show()
```

```python
##Create a box and whisker chart that showcases the price distribution of all listings split by room type.

price_roomtype = df[['room_type','price']]
price_roomtype
sns.boxplot(x='room_type', y='price', data = price_roomtype)
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.title('Price Distribution of Listings by Room Type')
plt.show()
```
